```python
import pandas as pd
import re
```

```python
def all_html_tags(text):
    pattern = r'</?\s*([a-zA-Z0-9]+)'
    return re.findall(pattern,text)

```

```python
text_data = pd.read_csv("IMDB Dataset.csv")

```

```python
text_data["html_tags"]=text_data["review"].apply(all_html_tags)
text_data["unique_html_tags"] = text_data["html_tags"].apply(set)
```

```python
other_tags = text_data[
    text_data["unique_html_tags"].apply(
        lambda tags: (len(tags) > 1) or (len(tags)==1 and "br" not in tags)
    )
]
```

```python
unique_tags = set(tag for tags in text_data['html_tags'] for tag in tags)
print(unique_tags)
```

```python
text_data["review"] = text_data["review"].str.replace(
    r'<br\s*/?>',
    ' ',
    regex=True,
    case=False
)

text_data["review"] = text_data["review"].str.replace(
    r'\s+',
    ' ',
    regex=True
).str.strip()
```

```python
text_data["html_tags"]=text_data["review"].apply(all_html_tags)
text_data["unique_html_tags"] = text_data["html_tags"].apply(set)
```

```python
other_tags = text_data[
    text_data["unique_html_tags"].apply(
        lambda tags: (len(tags) > 0)
    )
]
```

```python
def tokenize(text):
    lines = re.split(r'\.+', text.lower())
    return [
        re.findall(r"[a-z0-9']+", line)
        for line in lines
        if line.strip()
    ]
```

```python
text_data["tokens"]=text_data["review"].apply(tokenize)
```

```python
all_sentences = [
    sentence
    for review in text_data["tokens"]
    for sentence in review
]
```
